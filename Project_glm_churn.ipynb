```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("churn.csv")
data.head()
```

```python
data.columns
```

```python
# รายชื่อคอลัมน์ที่ต้องการแปลง
columns_to_convert = ['churn', 'internationalplan', 'voicemailplan']
data['churn'].replace({'No': 0, 'Yes': 1}, inplace=True)
data['churn'].replace({'no': 0, 'yes': 1}, inplace=True)
data['internationalplan'].replace({'no': 0, 'yes': 1}, inplace=True)
data['voicemailplan'].replace({'no': 0, 'yes': 1}, inplace=True)
```

```python
data
```

```python
data['churn'].dtype
```

เปลี่ยนเป็น int

```python
data['churn'] = data['churn'].astype(int)
data['internationalplan'] = data['internationalplan'].astype(int)
data['voicemailplan'] = data['voicemailplan'].astype(int)
```

#Scatter plot

```python
numeric_vars = ['accountlength', 'internationalplan', 'voicemailplan',
                'numbervmailmessages', 'totaldayminutes', 'totaldaycalls',
                'totaldaycharge', 'totaleveminutes', 'totalevecalls', 'totalevecharge',
                'totalnightminutes', 'totalnightcalls', 'totalnightcharge',
                'totalintlminutes', 'totalintlcalls', 'totalintlcharge',
                'numbercustomerservicecalls', 'churn']

# สร้าง Scatter Plot สำหรับแต่ละตัวแปร
for var in numeric_vars:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=var, y='churn', data=data, hue='churn', palette='viridis', marker='o')
    plt.title(f'Scatter Plot of {var} by Churn')
    plt.xlabel(var)
    plt.ylabel('Churn')
    plt.show()
```

# Corelation

```python
c_map=sns.color_palette("magma", as_cmap=True)
    # c_map color map sns ในตัวอย่างใช้ magma
d_correlation = data.corr(numeric_only=True).round(2)
    # d_correlation บงบอกชุดข้อมูลที่จะนำมาหา cor
        # numeric_only=True = ตัวเลขเท่านั้น
plt.figure(figsize = (12,8))
    # ขนาด12*8นิ้ว
sns.heatmap(d_correlation, annot = True, cmap = c_map)
plt.show()
```

เตรียมตัวแปรก่อนทำML

```python
df = data[['churn','internationalplan','totaldayminutes','totaldaycharge','numbercustomerservicecalls']]
```

# ML

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop('churn',axis=1)
Y = df['churn']

x = df.drop('churn',axis=1)
y = df['churn']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
print(f'X_train data size: {X_train.shape}')
print(f'Y_train data size: {Y_train.shape} \n')

print(f'X_test data size : {X_test.shape}')
print(f'Y_test data size : {Y_test.shape} \n')

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# แบ่งข้อมูลเป็น train set และ test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(f'X_train data size: {x_train.shape}')
print(f'Y_train data size: {y_train.shape} \n')
print(f'X_test data size : {x_test.shape}')
print(f'Y_test data size : {y_test.shape} \n')

# สร้างและฝึกโมเดลโดยไม่ให้เพิ่ม intercept term
model = LogisticRegression(fit_intercept=False, random_state=42)
model.fit(x_train, y_train)

# ทำนายผล
y_pred = model.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')
```

```python
import statsmodels.api as sm

# สร้างและฝึกโมเดล
model = sm.Logit(y_train, sm.add_constant(x_train)).fit()

# ดึง summary ของโมเดล
summary = model.summary()
print(summary)
```
