```python
import pandas as pd
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
wine
```

```python
#print(wine['data'])  # ข้อมูลเกี่ยวกับ features หรือคุณสมบัติของไวน์
#print(wine['target'])  # ข้อมูลเกี่ยวกับ target class หรือคลาสของไวน์
#print(wine['target_names'])  # ชื่อของคลาส
#print(wine['feature_names'])  # ชื่อของ features
#print(wine['DESCR'])  # คำอธิบายเกี่ยวกับ dataset
```

```python
print(wine['DESCR'])  # คำอธิบายเกี่ยวกับ dataset
```

```python
X = wine.data
y = wine.target
features = wine.feature_names
print(len(features))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for i in range(13):
    sns.boxplot(x=y, y=X[:,i])
    plt.ylabel(features[i])
    plt.show()
```

```python
for i in range(13):
    plt.hist(X[:,i], edgecolor='black')
    plt.title(features[i])
    plt.show()
```

```python
target_name = wine.target_names
print(target_name)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 32)
```

```python
kernel = ['linear', 'rbf', 'poly']
for i in kernel:
    model = SVC(kernel=i , C=1.0)
    model.fit(X_train, y_train)
    print('For kernel:' , i)
    print('Accuracy is:' ,model.score(X_test, y_test))
```

```python
model = SVC()
model.fit(X_train, y_train)

print("Accuracy On Trainingset is : ", model.score(X_train, y_train))
print("Accuracy On Testingset is : ", model.score(X_test, y_test))
```

```python
# มีเเค่ poly อย่างเดียวที่ต้องกำหนด degree linear, rbf ไม่ต้องใส่
for i in range(1,10):
    model=SVC(kernel = 'poly', degree=i, C=100)
    model.fit(X_train, y_train)
    print("i is: ",i ,"Accuracy On Trainingset is : ", model.score(X_train, y_train))
    print("i is: ",i ,"Accuracy On Testingset is : ", model.score(X_test, y_test))
```

Linear, rbf, poly, sigmoid

```python
kernel = ['linear', 'rbf', 'poly','sigmoid']
for i in kernel:
    model = SVC(kernel=i , C=1.0)
    model.fit(X_train, y_train)
    print('For kernel:' , i)
    print('Accuracy is:' ,model.score(X_test, y_test))
```

## *GridSearchCV*

```python
from sklearn.model_selection import GridSearchCV

param_grid ={ "C" :[0.1, 1, 100 ,1000], 'kernel':['rbf', 'poly', 'sigmoid', 'linear'], 'degree':[1, 2, 3, 4, 5, 6] }
grid=GridSearchCV(SVC(),param_grid)
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.score(X_test, y_test))
```

## Report , Score , Classification

```python
from sklearn.metrics import classification_report
```

```python
model = SVC(kernel='linear' , C=1.0)
model.fit(X_train, y_train)
# ทำนายค่า target จากชุดข้อมูล X_test
y_pred = model.predict(X_test)

# สร้าง classification report
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# สร้าง Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# สร้างรูปภาพ
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g')

plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
