```python
import pandas as pd
 

superstore = pd.read_csv('Sample_Superstore.csv', encoding='windows-1252', index_col="Row ID") 
        #วิธีเเก้มาจาก stackoverflow ('utf-8' codec can't decode byte 0xa0 in position 4276: invalid start byte)
        # index_col => optional
superstore.head()
```

```python
superstore['Order Date'] = pd.to_datetime(superstore["Order Date"])
superstore['Ship Date'] = pd.to_datetime(superstore['Ship Date' ])
superstore['Postal Code'] = superstore['Postal Code'].astype(str)
#  ถ้าไม่มีจะ Error ตรง monthlySales
```

# 2 Product Monthlty Sales Trend
#### 2,1 Compute monthly sales

```python
# ทำการย้าย Column [Order Date] ไปอยู่หน้าสุด  (Order Date จะถูกเซ้ตให้กลายเป็น index ด้านหน้า)
superstore.set_index('Order Date').head(5)
```

```python
monthlySales = superstore.set_index('Order Date')\
    .groupby(['Category', 'Sub-Category'])['Sales']\
    .resample('M')\
    .sum()\
    .reset_index()

monthlySales.head()
```

#### 2.2 Convert Order Date to Period

```python
# ทำ order date ให้เป็นรายเดือน เพื่อให้ดูง่ายขึ้น
monthlySales["Order Date"] = monthlySales["Order Date"].dt.to_period('M')
monthlySales.head()
```

### 2.3 Cross tabulation

```python
monthlySales.pivot_table(
    index = ['Category', 'Sub-Category'],
    columns = 'Order Date',
            values = 'Sales', aggfunc=sum
)
```
