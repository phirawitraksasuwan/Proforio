# Aggregate Statistics(Group By)

```python
df = pd.read_csv('Pokemon.csv')
df.groupby(['Type 1']).mean().round(2).sort_values('Defense', ascending=False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1', 'Type 2']).count()['#'].reset_index()
```

# Working with large amounts of data

```python
# ไม่เข้าใจ
df_new = pd.DataFrame(columns=df.columns)
for df in pd.read_csv('Pokemon.csv',chunksize=5):
    results = df.groupby(["Type 1"]).count()

df_new = pd.concat([df_new, results])
```
