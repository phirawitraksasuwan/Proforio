```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
df['Region'].unique()
region_sales = df[df["Order Date"].dt.year == 2019].groupby('Region')['Sales'].sum()
total_sales = df[df["Order Date"].dt.year == 2019]['Sales'].sum()
total_sales_region_2019 = ( (region_sales/total_sales)*100).round(2)
total_W_C_2019 = total_sales_region_2019[ ["West","Central"] ] .sum() 
print("The proportion of sales of West and Central is :", total_W_C_2019 ,"%")
```

```python
df_2019_to_2020 = df[ (df['Order Date'].dt.year >= 2019) & (df['Order Date'].dt.year <= 2020) ]
df_2019_to_2020
```

```python
df_ProductName_Sales_top10 = ( ( df_2019_to_2020.groupby('Product Name')['Sales'].sum() ).sort_values(ascending=False) ).head(10)
df_ProductName_count_top10 = ( ( df_2019_to_2020.groupby('Product Name')['Product Name'].count() ).sort_values(ascending=False) ).head(10)

print("Top 10 popular product in terms of number of order : \n", df_ProductName_Sales_top10 )
print("==========================================================")
print("Top 10 popular product in total sales : \n", df_ProductName_count_top10 )
```
